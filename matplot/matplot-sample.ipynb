```python
import matplotlib as mpl
import matplotlib.pylab as plt
```

```python
plt.title("Plot")
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
plt.title("x ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

```python
plt.title('Title')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
#스타일 문자열은 색깔(color), 마커(marker), 선 종류(line style)의 순서로 지정한다.
plt.title("'rs--' style plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title('Set Styles')
plt.show()
```

```python
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x,y ticks")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
plt.title("x axis, y axis range")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```
